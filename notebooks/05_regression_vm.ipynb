# 05 · 회귀 VM — 연속 계측값 예측 (정통 Virtual Metrology)

분류(양/불량)를 넘어 **막두께 같은 연속 계측값**을 센서로 예측. RMSE·R²로 평가.

> 아래는 **합성 증착 데이터**(방법 시연용). 네 XPS/박막 실측(공정조건→막두께)으로 교체하면 그대로 포트폴리오.

```python
import sys; sys.path.append("..")
import numpy as np, matplotlib.pyplot as plt
from src.regression import make_synthetic_vm, evaluate, vm_pipeline, PLSReg
X, y = make_synthetic_vm(n=1200, n_sensors=40)
print("X:", X.shape, "| thickness range:", round(float(y.min()),1), "~", round(float(y.max()),1), "nm")
```

## 모델별 회귀 성능 (5-fold CV)

```python
evaluate(X, y, cv=5)
```

## Parity plot — 예측 vs 실제 (RandomForest)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_predict
pred = cross_val_predict(vm_pipeline(RandomForestRegressor(n_estimators=400, n_jobs=-1, random_state=42)), X, y, cv=5)
plt.scatter(y, pred, s=8, alpha=.4, color="#1A4D8F")
lim = [y.min(), y.max()]; plt.plot(lim, lim, "r--")
plt.xlabel("true thickness (nm)"); plt.ylabel("predicted"); plt.title("VM parity — rf"); plt.show()
```
