결측 제거 → 중앙값 대체 → 분산 임계 → 스케일. 차원 변화 확인.

```python
import sys; sys.path.append("..")
import numpy as np, pandas as pd
from src.data import load_secom
from src.features import build_preprocessor
X, y, ts = load_secom()
print("raw:", X.shape, "| missing(mean):", round(float(X.isna().mean().mean()), 4))
```

## 전처리 적용 (fit은 train에서만 해야 누수 없음 — 여기선 데모로 전체 fit)

```python
pre = build_preprocessor().fit(X)
Xp = pre.transform(X)
print("after preprocessing:", Xp.shape)
dropped_missing = int((~pre.named_steps["drop_missing"].keep_).sum())
dropped_const = int((~pre.named_steps["var"].get_support()).sum())
print("dropped (>50% missing):", dropped_missing)
print("dropped (constant)    :", dropped_const)
print("surviving features    :", Xp.shape[1])
```

> 핵심: 실제 CV에서는 이 `pre`를 모델과 함께 `Pipeline`으로 묶어 **fold마다 train에서만 fit** → 검증 누수 0.
